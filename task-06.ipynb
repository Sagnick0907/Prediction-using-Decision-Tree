## Importing Packages ##

First, let's import all the packages that will be needed during this assignment. 

```python
import sklearn.datasets as datasets
import pandas as pd
import seaborn as sns
import sklearn.metrics as sm
```

## Overview of the Problem set ##

**Problem Statement**: 
Create the Decision Tree classifier and visualize it graphically.

```python
# Importing Dataset

iris=datasets.load_iris()
Data = pd.DataFrame(iris.data, columns=iris.feature_names)
print("Data is successfully imported!")
```

```python
# Printing the first 5 rows of dataset...
Data.head()
```

```python
# Printing the last 5 rows of dataset...
Data.tail()
```

```python
# descibe() helps us to see the percentiles, mean, std, max, count of the given dataset.
Data.describe()
```

```python
# Printing the full summary of the dataframe...
Data.info()
```

```python
Data.isnull().sum()
```

```python
Class = iris.target
Class
```

## Splitting Dataset into Train and Test Sets ## 



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Data, Class, test_size=0.33, random_state=42)
```

## Defining the Decision Tree Algorithm


```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)

print('Decision Tree Classifer Created Successfully')
```

```python
y_predict = dtc.predict(X_test)
```

## Constructing Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

```python
print("Accuracy of the Model:",sm.accuracy_score(y_test,y_predict))
```

The accuracy is 1 or 100% since i took all the 4 features of the iris dataset.

## Visualizing the Decision Tree


```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
fn=['Sepal length (cm)','Sepal width (cm)','Petal length (cm)','Petal width (cm)']
cn=['Setosa','Versicolor','Virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtc, feature_names = fn, class_names = cn, filled = True);
```
